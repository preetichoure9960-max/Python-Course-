```python
# normal way
squares = []
for i in range(1, 6):
    squares.append(i*i)

# shortcut
squares = [i*i for i in range(1, 6)]

print(squares)
```

```python
cubes = []
for i in range(1, 6):
    cubes.append(i*i*i)

cubes = [i*i*i for i in range(1, 6)]

print(cubes)
```

```python
nums = [1, 2, 3, 4, 5]

even = odd = 0

for i in nums:
    if i % 2 == 0:
        even += 1
    else:
        odd += 1

print("Even:", even)
print("Odd:", odd)
```

```python
nums = [10, 50, 30]

largest = nums[0]

for i in nums:
    if i > largest:
        largest = i

print("Largest:", largest)
```

```python
nums = [1, 2, 3]

rev = []
for i in range(len(nums)-1, -1, -1):
    rev.append(nums[i])

print(rev)
```

```python
nums = [1, 2, 3, 2, 1]

if nums == nums[::-1]:
    print("Palindrome")
else:
    print("Not Palindrome")
```
